```python
import chromadb 
from chromadb.config import Settings

path = '/Users/jordanjones/Documents/sherlock/houn.txt'
import pandas as pd
import numpy as np

def preprocess(data):

    import re

    def clean_text(document):
        document = document.lower() #lower case
        document = re.sub('\n+', ' ', document ) #Get rid of the string \n
        document = re.sub("'", "", document) #Prevent don't becoming don t etc
        document = re.sub('[^.\?\!\w\s+]', ' ', document) #Remove anything that is not digits or letters and then remove whitespace
        document = re.sub(r'\s+', ' ', document) #Reomove all whitepaces
        
        return document
    data = clean_text(data)



    from nltk.tokenize import sent_tokenize
    text = sent_tokenize(data)

    
    return text
```

```python
with open(path) as f:
    data = f.read()
        

x = preprocess(data)

print(len(x))

x=x[1:3847]
ids = [f'sentence_{i}' for i in range(len(x))] #F string sentence_itterable 

print(x[0])
print(x[-1])
```

```python
Sherlock = chromadb.PersistentClient(path='db')
print('done')

collection3 = Sherlock.create_collection(name='sherlock_sentences7')



from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-MiniLm-L6-v2')
texts = x
story_embeddings = model.encode(texts)
print(story_embeddings)

collection3.add(embeddings = story_embeddings,
               documents=texts,
               ids = ids)
```

```python

results = collection3.query(query_texts=['find sentences related to sherlock'
], n_results=5)
results
```
